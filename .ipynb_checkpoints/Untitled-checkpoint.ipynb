```python
#Discuss string slicing and provide examples?

'''
String slicing in Python is a method for extracting parts of a string based on specified indices.
It's a versatile feature that allows you to handle substrings, reverse strings, and more, with concise syntax. 

The syntax for string slicing in Python is:
string[start:end:step]

start: The starting index of the slice (inclusive).
end: The ending index of the slice (exclusive).
step: The step or stride between indices.
'''

s = "Hello, World!"
print(s[0:5])  # Output: Hello
# This extracts the substring from index 0 up to, but not including, index 5.

s = "Hello, World!"
print(s[:5])   # Output: Hello
print(s[7:])   # Output: World!

#s[:5] extracts from the beginning to index 5 (excluding index 5).
#s[7:] extracts from index 7 to the end of the string.

s = "Hello, World!"
print(s[::2])  # Output: Hlo ol!
#s[::2] extracts every second character from the string.

s = "Hello, World!"
print(s[-6:-1])  # Output: World

#Negative indices count from the end of the string. Here, s[-6:-1] extracts from 
#the 6th last character up to, but not including, the 1st last character.

s = "Hello, World!"
print(s[::-1])  # Output: !dlroW ,olleH

#A negative step reverses the string. s[::-1] reverses the entire string.

s = "Hello, World!"
print(s[20:])   # Output: (empty string)
print(s[:-20])  # Output: (empty string)

#When indices are out of bounds, Python handles it gracefully by returning an empty string if the slice is invalid.

s = "abcdefghijk"
print(s[1:8:2])  # Output: bdfh

#Here, s[1:8:2] extracts characters from index 1 to 7 with a step of 2, giving every second character in that range.

s = "Python"
print(s[5:1:-1])  # Output: noht

#This extracts characters from index 5 to 2 in reverse order (-1 step), resulting in noht.
```

```python
#Explain the key features of lists in Python
'''Lists in Python are one of the most versatile and commonly used data structures.
They allow you to store and manage collections of items efficiently.'''

# 1. Ordered Collection: Lists maintain the order of elements as they were added. 
#This means that the order in which you append elements to a list is preserved.

my_list = [10, 20, 30]
print(my_list[0])  # Output: 10

#2. Mutable:Lists can be modified after creation. You can change, add, or remove elements.

my_list = [1, 2, 3]
my_list[1] = 20
my_list  # Output: [1, 20, 3]

#3. Dynamic Sizing: Lists can grow or shrink in size as needed. You don't have to specify the size when you create a list

my_list = [1, 2]
my_list.append(3)
print(my_list)  # Output: [1, 2, 3]

#Heterogeneous Elements:Lists can store elements of different data types, including other lists.

my_list = [1, "hello", 3.14, [1, 2, 3]]
print(my_list[3])  # Output: [1, 2, 3]

#5. Indexing and Slicing: Lists support indexing to access individual elements and slicing to access a subset of elements.

my_list = [0, 1, 2, 3, 4]
print(my_list[1])   # Output: 1
print(my_list[1:4]) # Output: [1, 2, 3]

#6. Iterable: Lists are iterable, which means you can use loops (like for loops) to iterate over their elements.

my_list = [1, 2, 3]
for item in my_list:
    print(item)
# Output:
# 1
# 2
# 3

#7. Dynamic Methods : Lists come with a variety of built-in methods for adding, removing, and manipulating elements.
'''
append(item): Adds an item to the end.
extend(iterable): Extends the list by appending elements from an iterable.
insert(index, item): Inserts an item at a given position.
remove(item): Removes the first occurrence of an item.
pop(index): Removes and returns the item at the given position.
sort(): Sorts the list in ascending order.
reverse(): Reverses the elements of the list in place.'''

my_list = [3, 1, 2]
my_list.append(4)
my_list.insert(1,5)
my_list.remove(5)
my_list.pop(1)
my_list.sort()
print(my_list) 

#8. List Comprehensions: Python supports list comprehensions, a concise way to create lists based on existing lists.

squares = [x**2 for x in range(5)]
print(squares)  # Output: [0, 1, 4, 9, 16]

#9. Nested Lists: Lists can contain other lists, which allows for the creation of multi-dimensional arrays.

matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
print(matrix[1][2])  # Output: 6

#10. Supports Operations: Lists support various operations such as concatenation, repetition, and membership testing.

list1 = [1, 2]
list2 = [3, 4]
print(list1 + list2)  # Output: [1, 2, 3, 4]
print(list1 * 3)      # Output: [1, 2, 1, 2, 1, 2]
print(2 in list1)    # Output: True



```

```python
# Describe how to access modify and delete elements in a list with examples

#1. Accessing Elements

my_list = ['apple', 'banana', 'cherry']
my_list[1]     # Output: banana
my_list[-2]  # Output: banana
my_list[1:3]  # Output: ['banana', 'cherry']
my_list[:2]   # Output: ['apple', 'banana']

#2. Modifying Elements:

my_list = [1, 2, 3, 4]
my_list[2] = 99
print(my_list)  # Output: [1, 2, 99, 4]

my_list = [1, 2, 3, 4, 5]
my_list[1:3] = [20, 30]
print(my_list)  # Output: [1, 20, 30, 4, 5]

my_list = [1, 2, 4]
my_list.insert(2, 3)  # Insert 3 at index 2
print(my_list)  # Output: [1, 2, 3, 4]

#3. Deleting Elements

my_list = ['a', 'b', 'c', 'd']
del my_list[1]  # Remove element at index 1
print(my_list)  # Output: ['a', 'c', 'd']

my_list = ['apple', 'banana', 'cherry']
my_list.remove('banana')
print(my_list)  # Output: ['apple', 'cherry']

my_list = [10, 20, 30, 40]
removed_element = my_list.pop(2)  # Remove and return element at index 2
print(removed_element)  # Output: 30
print(my_list)  # Output: [10, 20, 40]

my_list = [1, 2, 3, 4]
my_list.clear()
print(my_list)  # Output: []
```

```python
# Compare and contrast tuples and lists with examples

'''Tuples and lists are both fundamental data structures in Python used to store collections of items.
However, they have distinct differences and use cases'''

# 1. Mutability

'''Lists: Lists are mutable, meaning you can change their content after they are created.
This includes adding, removing, or modifying elements.'''

my_list = [1, 2, 3]
my_list[1] = 20
my_list.append(4)
print(my_list)  # Output: [1, 20, 3, 4]

'''Tuples: Tuples are immutable, meaning once they are created, their content cannot be changed.
You cannot add, remove, or modify elements.'''

my_tuple = (1, 2, 3)
# my_tuple[1] = 20  # This will raise a TypeError
# my_tuple.append(4)  # This will raise an AttributeError
print(my_tuple)  # Output: (1, 2, 3)


#2. Syntax

#Lists: Defined with square brackets [].
my_list = [1, 2, 3]
my_list

#Tuples: Defined with parentheses ().
my_tuple = (1, 2, 3)
my_tuple

#3. Methods and Operations

#Lists: Have many built-in methods for modification, such as append(), extend(), insert(), remove(), pop(), and clear().
my_list = [1, 2, 3]
my_list.append(4)
my_list.remove(2)
print(my_list)  # Output: [1, 3, 4]

#Tuples: Have fewer built-in methods, mainly count() and index(). They don’t support methods for modification
my_tuple = (1, 2, 3)
print(my_tuple.count(2))  # Output: 1
print(my_tuple.index(3))  # Output: 2

#4. Use Cases

'''Lists: Suitable for collections of items where the content may change over time. Lists are ideal when you need 
a sequence of items that you might need to update or reorder.'''

shopping_list = ['milk', 'eggs', 'bread']
shopping_list.append('butter')  # Adding an item
shopping_list[1] = 'cheese'     # Modifying an item

'''Tuples: Suitable for fixed collections of items that should not change. 
Tuples are often used for representing records or fixed data structures where immutability is desired.'''

coordinates = (10.0, 20.0)  # Represents a fixed point in 2D space
person = ('Alice', 30, 'Engineer')  # Represents a person with a fixed set of attributes

#5. Hashability

#Lists: Not hashable, so they cannot be used as keys in dictionaries or elements in sets.
my_list = [1, 2, 3]
# my_set = set()
# my_set.add(my_list)  # This will raise a TypeError

#Tuples: Hashable (if they contain only hashable elements), so they can be used as keys in dictionaries or elements in sets.

my_tuple = (1, 2, 3)
my_set = set()
my_set.add(my_tuple)  # This works
print(my_set)  # Output: {(1, 2, 3)}

```

```python
# Describe the key features of sets and provide examples of their use

'''Sets in Python are a built-in data structure designed for storing unique items in an unordered manner.
They offer several key features and are useful for various tasks in programming.'''

#1. Unique Elements:Sets automatically enforce uniqueness among their elements. 
#Duplicate items are not allowed and are removed upon insertion.

my_set = {1, 2, 3, 3, 4}
print(my_set)  # Output: {1, 2, 3, 4}


#2. Unordered Collection: Sets do not maintain any specific order for their elements. 
#The order in which elements are displayed can vary.

my_set = {4, 1, 3, 2}
print(my_set)  # Output: {1, 2, 3, 4} (order may vary)

#3. Mutable: you can add or remove elements from a set after its creation. However, 
#the elements themselves must be immutable (e.g., numbers, strings, tuples).

my_set = {1, 2, 3}
my_set.add(4)
my_set.remove(2)
print(my_set)  # Output: {1, 3, 4}

#4. No Indexing:Sets do not support indexing, slicing, or other sequence-like behavior.
#To access elements, you must use iteration or membership tests.

my_set = {1, 2, 3}
# print(my_set[0])  # This will raise a TypeError

#5. Set Operations:Sets support mathematical set operations such as union, intersection, difference, and symmetric difference.

set_a = {1, 2, 3}
set_b = {3, 4, 5}

union_set = set_a | set_b           # Union
intersection_set = set_a & set_b    # Intersection
difference_set = set_a - set_b      # Difference
sym_diff_set = set_a ^ set_b        # Symmetric Difference

print(union_set)           # Output: {1, 2, 3, 4, 5}
print(intersection_set)    # Output: {3}
print(difference_set)      # Output: {1, 2}
print(sym_diff_set)        # Output: {1, 2, 4, 5}

#6. Membership Testing: You can test for the presence or absence of an element in a set using the in keyword.

my_set = {1, 2, 3}
print(2 in my_set)  # Output: True
print(4 in my_set)  # Output: False

#7. Set Comprehensions: Similar to list comprehensions, set comprehensions provide a concise way to create sets.
squares = {x**2 for x in range(5)}
print(squares)  # Output: {0, 1, 4, 9, 16}

#8. Frozensets: frozenset is an immutable variant of a set. It supports all set operations but cannot be modified after creation.

frozen_set = frozenset([1, 2, 3])
print(frozen_set)  # Output: frozenset({1, 2, 3})
# frozen_set.add(4)  # This will raise an AttributeError

```

```python
# Discuss the use cases of tuples and sets in Python programming


'''Tuples and sets are distinct data structures in Python, each with unique characteristics that make them suitable for 
different use cases.
Here's  use cases for tuples and sets:'''

#1. tuples:

#Fixed Collections: When you have a fixed collection of items that should not change, use a tuple. For example,
#representing coordinates (x, y) or a record of a person's details (name, age, address).

#Function Returns: Tuples are often used to return multiple values from a function. 
#For instance, a function might return (result, status_code).

#Unpacking: Tuples are useful for unpacking multiple values into individual variables. For example:
point = (3, 5)
x, y = point

#As Dictionary Keys: Tuples can be used as keys in dictionaries when the elements are immutable. 
#For instance, you might use a tuple of coordinates as a key in a grid-based game.

#Data Integrity: Tuples are used to ensure data integrity when you want to guarantee that the data structure remains unchanged.

#2. sets


#Membership Testing: Sets are optimized for membership tests. Checking if an item is in a set is generally faster than checking
#if an item is in a list or tuple due to the underlying hash table implementation.

#Eliminating Duplicates: If you need to remove duplicates from a collection, converting it to a set is a straightforward way 
#to achieve this:

items = [1, 2, 2, 3, 4, 4]
unique_items = set(items)  # {1, 2, 3, 4}

#Set Operations: Sets support mathematical operations like union, intersection, difference, and symmetric difference,
#which can be useful in various applications:

set1 = {1, 2, 3}
set2 = {3, 4, 5}
union = set1 | set2           # {1, 2, 3, 4, 5}
intersection = set1 & set2    # {3}
difference = set1 - set2      # {1, 2}

#Unique Element Tracking: When you need to track unique elements or ensure that elements appear only once, sets are a natural choice.

#Fast Lookup: For scenarios where you need fast lookups and membership checking, sets are preferred over lists or tuples
#due to their hash-based implementation.
```

```python
#Describe how to add, modify, and delete items in a dictionary with examples

#dictionaries are mutable, meaning you can add, modify, and delete items. Here’s how you can perform these operations:

'''Adding Items
To add an item to a dictionary, you can simply assign a value to a new key. If the key does not exist, it will be created.'''

my_dict = {'name': 'xyz', 'age': 30}

my_dict['city'] = 'india'

print(my_dict)
# Output: {'name': 'Alice', 'age': 30, 'city': 'New York'}

'''Modifying Items
To modify an item, you assign a new value to an existing key. This updates the value associated with that key.'''

my_dict = {'name': 'xyz', 'age': 30, 'city': 'india'}

my_dict['age'] = 31

print(my_dict)
# Output: {'name': 'Alice', 'age': 31, 'city': 'New York'}

'''Deleting Items
You can delete items from a dictionary using the del statement or the pop method.

Using del:

The del statement removes a key-value pair from the dictionary.'''

my_dict = {'name': 'xyz', 'age': 30, 'city': 'india'}

del my_dict['city']

print(my_dict)
# Output: {'name': 'Alice', 'age': 30}

'''Using pop:

The pop method removes the key-value pair and returns the value associated with the key. 
If the key is not found, it raises a KeyError, but you can provide a default value to avoid this.'''

my_dict = {'name': 'xyz', 'age': 30, 'city': 'india'}

removed_value = my_dict.pop('age')

print(my_dict)
# Output: {'name': 'Alice', 'city': 'New York'}
print('Removed value:', removed_value)
# Output: Removed value: 30

removed_value = my_dict.pop('country', 'Not Found')
print('Removed value with default:', removed_value)
# Output: Removed value with default: Not Found


```

```python
# Discuss the importance of dictionary keys being immutable and provide examples

#dictionary keys must be immutable, which is crucial for maintaining the integrity and efficiency of the dictionary.

'''Importance of Immutable Dictionary Keys
Hashing Consistency:

Hash Table Implementation: Dictionaries in Python are implemented using hash tables. The key-value pairs are
stored in a way that allows for fast access using the key's hash value.

Consistency Requirement: For the hash table to work correctly, the hash value of a key must remain 
constant throughout its lifetime. If a key were mutable and its hash value changed, it could lead to 
inconsistencies where the key cannot be found in the hash table, even though it was initially added.

Data Integrity:
Predictable Behavior: Immutable keys ensure that once a key is added to a dictionary, it cannot be altered 
in a way that would change its hash value. This predictability is essential for ensuring that operations like 
lookups, insertions, and deletions work reliably.

Performance:

Efficient Lookups: The immutability of keys allows Python to optimize dictionary operations. 
The hash value of immutable objects like integers and strings can be computed once and reused, 
leading to faster dictionary operations.'''

my_dict = {1: 'one', 2: 'two'}
print(my_dict[1])  # Output: 'one'

my_dict = {'name': 'xyz', 'age': 30}
print(my_dict['name'])  # Output: 'Alice'

my_dict = {('x', 'y'): 'coordinates'}
print(my_dict[('x', 'y')])  # Output: 'coordinates'
```
